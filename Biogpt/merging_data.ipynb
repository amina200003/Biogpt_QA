```python
#Creation d'un dataset qui fusionne les datasets medquad, medqaUSMLE,pubmedqa et medication QA
import pandas as pd
import os
import json
import chardet
# ------------ MEDQUAD ------------
df = pd.read_csv("medquad.csv")
df = df[["question", "answer", "source"]]
df_medquad = df.rename(columns={"question": "question", "answer": "answer", "source": "source"})

# ------------ MEDQA USMLE ------------
docs_medqa = []
path = "medqa_exam"
for nom_fichier in os.listdir(path):
    if nom_fichier.endswith(".jsonl"):
        chemin_fichier = os.path.join(path, nom_fichier)
        with open(chemin_fichier, 'r', encoding='utf-8') as f:
            for ligne in f:
                try:
                    data = json.loads(ligne)
                    if all(k in data for k in ["question", "answer"]):
                        docs_medqa.append({
                            "question": data["question"],
                            "answer": data["answer"],
                            "source": "National Medical Board Examination(USA)"
                        })
                except json.JSONDecodeError:
                    print(f"Erreur de d√©codage dans le fichier {nom_fichier} : {ligne}")
df_medqa = pd.DataFrame(docs_medqa)

#------------------------- PubMedQA ------------------
with open("pubmedqa.json") as f:
    pubmedqa_raw = json.load(f)

# for key,val in pubmedqa_raw.items():
#     print(key,"---",val)

docs_pubmedqa = []
for val in pubmedqa_raw.values():
              docs_pubmedqa.append({
              "question": val["QUESTION"],
              "answer": val["LONG_ANSWER"],
              "source": "PubMed"
          })
df_pubmedqa = pd.DataFrame(docs_pubmedqa)
#print(df_pubmedqa)


# ------------ MEDICATIONQA ------------
with open("medicationQA.csv", 'rb') as f:
    result = chardet.detect(f.read(10000))
df2 = pd.read_csv("medicationQA.csv", encoding=result["encoding"], on_bad_lines="skip", sep="\t")
df2.columns = df2.columns.str.strip()
df2 = df2[["Question", "Answer", "URL"]]
df_medication = df2.rename(columns={"Question": "question", "Answer": "answer", "URL": "source"})

# ------------ Fusion des datasets ------------
df_all = pd.concat([df_medqa, df_medquad, df_medication,df_pubmedqa], ignore_index=True)
df_all.dropna(subset=["question", "answer", "source"], inplace=True)
df_all = df_all[df_all["question"].str.strip() != ""]
df_all = df_all[df_all["answer"].str.strip() != ""]
df_all = df_all[df_all["source"].str.strip() != ""]
# ------------ Sauvegarde CSV ------------
df_all.to_csv("merged_medical_QA_dataset.csv", index=False, encoding="utf-8")
```
